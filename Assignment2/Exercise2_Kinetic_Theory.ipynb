```python
#%%time 

from vpython import *
import numpy as np
from scipy.constants import k # Boltzmann constant. 

# Hard-sphere gas.
# Bruce Sherwood

win = 500

Natoms = 100  # Amount of atoms in simulation. 

# Typical values
L = 1 # container is a cube with sides L.
gray = color.gray(0.7) # color of edges of container.
mass = 4E-3/6E23 # helium mass.
Ratom = 0.03 # wildly exaggerated size of helium atom.
T = 300 # around room temperature.
dt = 1E-5 # time interval for updating positions. 

# Animation-canvas setup. 
animation = canvas(width=win, height=win, align='left')
animation.range = L
animation.title = 'A "hard-sphere" gas (He)'
s = """ Theoretical and averaged speed distributions (meters/sec).
 Initially all atoms have the same speed, but collisions
 change the speeds of the colliding atoms. One of the atoms is
 marked and leaves a trail so you can follow its path.

"""
animation.caption = s

# Makes the confinement box for the gas. 
d = L/2+Ratom
r = 0.005
boxbottom = curve(color=gray, radius=r)
boxbottom.append([vector(-d,-d,-d), vector(-d,-d,d), vector(d,-d,d), vector(d,-d,-d), vector(-d,-d,-d)])
boxtop = curve(color=gray, radius=r)
boxtop.append([vector(-d,d,-d), vector(-d,d,d), vector(d,d,d), vector(d,d,-d), vector(-d,d,-d)])
vert1 = curve(color=gray, radius=r)
vert2 = curve(color=gray, radius=r)
vert3 = curve(color=gray, radius=r)
vert4 = curve(color=gray, radius=r)
vert1.append([vector(-d,-d,-d), vector(-d,d,-d)])
vert2.append([vector(-d,-d,d), vector(-d,d,d)])
vert3.append([vector(d,-d,d), vector(d,d,d)])
vert4.append([vector(d,-d,-d), vector(d,d,-d)])

# Oppgave 2a: Fyll inn!
pavg = 1 # average kinetic energy p**2/(2mass) = (3/2)kT.

def initiate_atoms(**kwargs):
    """Initiate atom-shapes, positions and their momenta."""
    Atoms = []
    p = []
    apos = []   

    for i in range(Natoms):
        x = L*random()-L/2 #random() returns a pseudo-random number between 0 and 1.     
        y = L*random()-L/2
        z = L*random()-L/2
        if i == 0:
            Atoms.append(sphere(pos=vector(x,y,z), 
               radius=Ratom, color=color.cyan, make_trail=True, retain=20, trail_radius=0.3*Ratom))
        else: 
            Atoms.append(sphere(pos=vector(x,y,z), radius=Ratom, color=gray))
        apos.append(vector(x,y,z))
        theta = pi*random()
        phi = 2*pi*random()
        px = pavg*sin(theta)*cos(phi)
        py = pavg*sin(theta)*sin(phi)
        pz = pavg*cos(theta)
        p.append(vector(px,py,pz))
    return Atoms, p, apos

Atoms, p, apos = initiate_atoms(Ratom = Ratom, Natoms = Natoms, L = L, pavg = pavg)

deltav = 100 # binning for v histogram.

def barx(v):
    """Index speed v into array of bars."""
    return int(v/deltav) 

xmax = 3000 # Maximum x-value in graph. 
nhisto = barx(4500) # Number of histogram bars.  
histo = np.zeros(nhisto) # Makes array for histogram bars. 
histo[barx(pavg/mass)] = Natoms # All atoms start at speed vavg = pavg/mass. 

# Make graph. 
gg = graph(width=win, height=0.4*win, xmax=xmax, align='left',
    xtitle='speed, m/s', ytitle='Number of atoms', ymax=Natoms*deltav/1000)


dv = 10
data = []
for v in range(0,xmax+dv,dv): # theoretical prediction
    # Oppgave 2a: Bytt ut strengen nedenfor med Maxwells fordeling. 
    data.append([v, (deltav*Natoms)*"Bytt meg ut!"])

theory = gcurve(color=color.cyan, data = data) # Makes curve for theoretical prediction. 
theory.data = data # Needs to be set manually here also, to be able to get theory.data later.

# Initialize bins for histogram in list. 
accum = [] 
for i in range(barx(xmax)): 
    accum.append([deltav*(i+.5),0])

    
# Makes histogram bars. 
vdist = gvbars(color=color.red, delta=deltav)

def interchange(v1, v2): 
    """Remove 1 speed from v1 and add 1 speed to v2."""
    barx1 = barx(v1)
    barx2 = barx(v2)

    # Check that function arguments make sense. 
    if barx1 == barx2:  
        return
    if barx1 >= len(histo) or barx2 >= len(histo): 
        return

    histo[barx1] -= 1
    histo[barx2] += 1

def check_collisions(**kwargs):
    """Check what atoms have collided. Returns a list of atoms that have collided."""
    hitlist = []
    d = 2*Ratom
    for i in range(Natoms):
        ai = apos[i]
        for j in range(i):
            # Oppgave 2a: Fullfør denne kodesnutten.  
            aj = apos[i]
            
            #Noen tips nedenfor: 
            
            # Finn vektoren mellom ai og aj.
            # mag2(vec) gir vec sin kvadrerte lengde. 
            # Hvis de har kollidert skal begge indeksene legges til i hitlist slik: 
            # hitlist.append([i, j])
    return hitlist

nhisto = 0 # number of histogram snapshots to average. 

def redirect_at_walls(**kwargs):
    """Check if atoms have hit the walls and changes the momenta accordingly. 
    
    Returns pressure in gas, counted from amount of collisions with surface. 
    """
    pressure = 0
    area = 6*L**2
    
    for i in range(Natoms):
        loc = apos[i]
        
        if abs(loc.x) > L/2:
            if loc.x < 0: 
                p[i].x =  abs(p[i].x) 
            else: 
                pass
                # Fyll inn!
                
            pressure += 2*abs(p[i].x)/(dt*area) # Change in wall momentum is 2*p.
    
        if abs(loc.y) > L/2:
            if loc.y < 0: 
                pass 
                # Fyll inn!
            else: 
                pass
                # FYll inn!
                
            pressure += 2*abs(p[i].y)/(dt*area) # Change in wall momentum is 2*p.
        
        if abs(loc.z) > L/2:
            if loc.z < 0: 
                pass
                # Fyll inn!
            
            else: 
                pass
                # Fyll inn!
                
            pressure += 2*abs(p[i].z)/(dt*areas) # Change in wall momentum is 2*p.

    return pressure # Counted the pressure caused by the collisions with the wall. 

pressures = np.array([]) # Used to store pressures returned from redirect_at_walls() in each iteration. 

previous_steps = [] # Used to store accum-values in each iteration. 
step = 1 # Record the current step in each iteration of the loop. 
relative_step = 3 # The relative step for checking convergence between steps.
# Convergence is checked between every 'relative_step' steps. 

while True:
    rate(300) # Limit the number of calculations. rate(x) => Max x executions of loop per second. 
    # If the simulation is 'stuttering', you can try changing this. 
    # Keep it as the same value throughout task 3 though, since runtime depends on this. 
    
    # Accumulate and average histogram snapshots.
    for i in range(len(accum)): 
        accum[i][1] = (nhisto*accum[i][1] + histo[i])/(nhisto+1)

    # Limits the amount of times the histogram data is updated. 
    if nhisto % 10 == 0:
        vdist.data = accum # Updates histogram data.  
    nhisto += 1
    
    # Stop the program when the simulated values in this step, relative to a previous step, is "close enough", set by variable 'relative_step'. 
    if step > relative_step*1000: # Times 1000 (e.g) to avoid premature simulation stop.  
        if np.linalg.norm(np.array(accum)[:,1] - previous_steps[-relative_step]) < 1e-3: 
            print("The simulation is done")
            print("STEP", step)
            print("Simulated with "+str(Natoms)+" particles.")

            # Finn forventningsverdien til v**2 her!
            # Tips: accum lagrer dataen som plottes i histogram.  
            expected_value = 0

            print("Expected value:", expected_value)
            print("3*k*T/mass:", 3*k*T/mass)

            print("Pressure via ideal gas law:", Natoms*k*T/L**3)
            print("Found pressure via expected value:", 
                  (Natoms*mass*expected_value)/(3*L**3)) # p = N*m*expected_value/3V.

            pressure = sum(pressures)/len(pressures) # Calculates average of pressures. 
            print("Pressure from counted collisions (average):", pressure)

            print('\n') # Add a newline to give room for magic command %%time.
            break
    
    # Update all positions.
    for i in range(len(Atoms)): 
        Atoms[i].pos = apos[i] = apos[i] + (p[i]/mass)*dt

    # Check for collisions.
    hitlist = check_collisions(Natoms = Natoms, Ratom = Ratom, apos = apos)

    # Update momenta of the atoms that collided, according to hitlist. 
    for ij in hitlist:
        i = ij[0]
        j = ij[1]
        ptot = p[i]+p[j]
        posi = apos[i]
        posj = apos[j]
        vi = p[i]/mass
        vj = p[j]/mass
        vrel = vj-vi
        a = vrel.mag2
        if a == 0: 
            continue  # exactly same velocities.
        rrel = posi-posj
        if rrel.mag > Ratom: 
            continue # one atom went all the way through another.

        # theta is the angle between vrel and rrel:
        dx = dot(rrel, vrel.hat)       # rrel.mag*cos(theta).
        dy = cross(rrel, vrel.hat).mag # rrel.mag*sin(theta).
        # alpha is the angle of the triangle composed of rrel, path of atom j, and a line.
        # from the center of atom i to the center of atom j where atome j hits atom i:
        alpha = asin(dy/(2*Ratom)) 
        d = (2*Ratom)*cos(alpha)-dx # distance traveled into the atom from first contact.
        deltat = d/vrel.mag         # time spent moving from first contact to position inside atom.
        
        posi = posi-vi*deltat # back up to contact configuration.
        posj = posj-vj*deltat
        mtot = 2*mass
        pcmi = p[i]-ptot*mass/mtot # transform momenta to cm frame.
        pcmj = p[j]-ptot*mass/mtot
        rrel = norm(rrel)
        pcmi = pcmi-2*pcmi.dot(rrel)*rrel # bounce in cm frame.
        pcmj = pcmj-2*pcmj.dot(rrel)*rrel
        p[i] = pcmi+ptot*mass/mtot # transform momenta back to lab frame.
        p[j] = pcmj+ptot*mass/mtot
        apos[i] = posi+(p[i]/mass)*deltat # move forward deltat in time.
        apos[j] = posj+(p[j]/mass)*deltat
        interchange(vi.mag, p[i].mag/mass) # A.mag = The magnitude of vector A. 
        interchange(vj.mag, p[j].mag/mass)

    pressure = redirect_at_walls(Natoms = Natoms) # Call function to check collisions with walls. 
    pressures = np.append(pressures, pressure) # Append pressure to pressures-list. 
    step += 1
    previous_steps.append(np.array(accum)[:,1]) # used to check convergence between every 'relative_step' steps.
```

## Ekstra: Kuriositeter i Python.

Du har sikkert sett, her eller andre steder, at ordene \*args og/eller \*\*kwargs står som parametre i funksjoner. Dette er to typer parametre som er vanlige å bruke i Python dersom man ønsker at funksjonen skal kunne kalles med et variabelt antall argumenter. 

\*args brukes dersom du ønsker å ta inn et variabelt antall argumenter til funksjonen din, uten nøkkelord. Alle argumentene som tas inn via \*args tolkes av funksjonen som en liste av verdier. Dette kan brukes på følgende måte. 

```python
def use_args(*args):
    for x in args:
        print(x)
        
    print()
    
    # Get last argument via index
    print(args[-1])
    
use_args(1, 2, 3, 4, 5, "hei", "kult", "Termisk fysikk")

# The function may also be called like below. 
arguments = [1, 2, 3, 4, 5, "hei", "kult", "Termisk fysikk"]
print("\nDette gir samme output som ovenfor, men bruker en liste som allerede er definert:")
use_args(*arguments)
```

\*\*kwargs minner veldig om \*args, men i stedet for at den tolkes som en liste, tolkes den som en dictionary. Dermed bruker man \*\*kwargs når man ønsker et variabelt antall argumenter i en funksjon, der man i tillegg ønsker å få inn ett nøkkelord (keyword) per argument. Et eksempel følger. 

```python
def use_kwargs(**kwargs):
    for key, value in kwargs.items():
        print(key, value)
    
    print()
    
    # Get value via key
    print(kwargs["fysikk_emner"])
  
use_kwargs(first = "one", second = "two", third = "3", 
           fysikk_emner = ["termisk fysikk", "elmag", "mekanisk fysikk"],
          slutt = "c u l8er")

# The function may also be called like below. 
keyword_args = {
    "first": "one", "second": "two", "third": "3", 
    "fysikk_emner": ["termisk fysikk", "elmag", "mekanisk fysikk"], 
    "slutt": "c u l8er"
}
print("\nDette gir samme output som ovenfor, men bruker en dictionary som allerede er definert:")
use_kwargs(**keyword_args)
```

\*args og \*\*kwargs brukes masse i de fleste moduler i Python. Hvis du ser nøye etter i dokumentasjonen til numpy og matplotlib vil du finne disse her også. 

Det finnes en enkel måte å aksessere indeksen i en `for`-løkke. Dette kan gjøres ved å 'wrappe' en 'iterable', dvs 
det du vil iterere over, som for eksempel en liste eller en dictionary, med `enumerate`. Et eksempel følger.

```python
def enumerate_example(*args):
    for i, arg in enumerate(args):
        print(i, arg)
    
enumerate_example("One", "Two", "Three", "Four", "Last")
```
